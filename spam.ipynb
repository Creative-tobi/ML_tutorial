```python
import pandas as pd
import numpy as np

data = {
    "EmailID": [10, 20, 30, 40, 50, 60],
    "Subject": [
        "WIN $1000 NOW!!!",       # Loud Spam (3 bangs)
        "Meeting at 2pm",         # Normal
        np.nan,                   # Broken Data (Drop this)
        "Hello friend!!",         # Normal (2 bangs)
        "URGENT RESPONSE!!!",     # Loud Spam (3 bangs)
        "Lunch?"                  # Normal
    ],
    "Label": ["Spam", "Ham", "Spam", "Ham", "Spam", "Ham"]
}

df = pd.DataFrame(data)
print("Original Data:")
df
```

```python
df.dropna(subset=['Subject'])
```

```python
is_spam = {"Spam": 1, "Ham": 0}
df['Is_Spam'] = df['Label'].map(is_spam)
df
```

```python
df['Bang_Count'] = df["Subject"].str.count("!")
df
```

```python
filt = (df['Is_Spam'] == 1) & (df['Bang_Count'] > 2.0)
df.loc[filt]
```

```python
import pandas as pd

# Timestamps are in Milliseconds (ms)
data = {
    "Key": ["a", "s", "d", "f", "g"],
    "Press_Time":   [1000, 1200, 1500, 2000, 3500],
    "Release_Time": [1100, 1250, 1550, 2200, 4500], # Notice the last one took a long time!
    "User_Status":  ["Normal", "Normal", "Normal", "Normal", "Depressed"]
}

df = pd.DataFrame(data)
df
```

```python
df['Dwell_Time'] = (df['Release_Time'] - df['Press_Time'])
df
```

```python
fill = df['Dwell_Time'] > 500
df[fill]
```
